## Clean up `relig_income.csv` file using `pandas.melt()`

## `weather.csv` data using `pandas.melt()`

```python
# Load data
df2 = pd.read_csv('data/tidy_messy_data/weather.csv')
df2
```

## Mostly tidy data

```python
tidy_df2 = pd.melt(
    df2,
    id_vars=['id', 'year', 'month', 'element'],
    var_name='day',
    value_name='temperature'
)

# drop rows with missing temperature values
tidy_df2 = tidy_df2.dropna(subset=['temperature'])
tidy_df2.reset_index()
```

### Tidy data

```python
# Pivot the data to create separate columns for TMAX and TMIN
pivot_df = tidy_df2.pivot(
    index=['id', 'year', 'month', 'day'],
    columns='element',
    values='temperature'
).reset_index()

pivot_df
```
