# 1. Barplot
- A barplot shows the relationship between a numeric and a categorical variable.- 
The height of the bar usually represents its numeric valu.
- 
We often use error bars, but the simplest form represents only the mean or median of each da.a.


## Matplotlib
### A basic barplot
This example shows probably the most basic barplot you can do with python and matplotlib. The `bar()` function used with the following parameters:

- `x` : The x coordinates of the bars. (`x_pos` in the example)
- `height` : The height(s) of the bars.


(Source: https://python-graph-gallery.com/1-basic-barplot/)

```python
# Load libraries
import numpy as np
import matplotlib.pyplot as plt
 
# Create dataset
height = [3, 12, 5, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')
x_pos = np.arange(len(bars))
 
# Create bars
plt.bar(x_pos, height)
 
# Create names on the x-axis
plt.xticks(x_pos, bars)
 
# Show graphic
plt.show()
```

### Customize barplots

```python
# Load library
import pandas as pd

# Create a pandas DataFrame
height = [3, 12, 5, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')

df = pd.DataFrame({'height': height,
                   'bars': bars})
```

#### Change colors
First, let's try to change colors of all bars.

```python
# Set colors
color = (0.2, # redness
         0.4, # greenness
         0.2, # blueness
         0.6 # transparency
         ) 

# Create bars
fig, ax = plt.subplots()
ax.bar(df['bars'], df['height'],
       color=color)

# Show graph
plt.show()
```

If you want to give different color to each bar, just provide a list of color names to the color argument.

```python
# Use color names
color = ['lightblue', 'blue', 'purple', 'red', 'black']

# Create bars
fig, ax = plt.subplots()
ax.bar(df['bars'], df['height'],
       color=color)

# Show graph
plt.show()
```

The edgecolor argument allows you to color the borders of barplots. You can either put a color name like 'blue' or put a list of colors.

```python
color = ['lightblue', 'blue', 'purple', 'red', 'black']
edgecolor = ['red', 'orange', 'darkblue', 'darkred', 'yellow']

# Create bars
fig, ax = plt.subplots()
ax.bar(df['bars'], df['height'],
       color=color,
       edgecolor=edgecolor,
       linewidth=3)

# Show graph
plt.show()
```

## Seaborn
### A basic barplot
Let's start with a basic example showing the total bill paid by customers in 4 different days of the week. The `barplot()` function of the seaborn library is used here. The arguments passed to the function are:

- `x` : the name of the dataframe column used for the X axis.

## Matplotlib

```python
# Load libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

First, let's create an artificial dataset. We use `np.random.normal()` to create normally distributed data for 'Group A' and 'Group B', but with different means.

```python
# Generate data for Group A (Normal Distribution)
group_a = np.random.normal(loc=50, scale=10, size=1000)

# Generate data for Group B (Skewed Distribution)
group_b = np.random.normal(loc=20, scale=10, size=1000)

# Create a DataFrame
df = pd.DataFrame({'Group': ['Group A'] * 1000 + ['Group B'] * 1000,
                   'Value': np.concatenate((group_a, group_b))})
```

Once we've opened our dataset, we'll now create the graph. The following displays the distribution of both groups using the `boxplot()` function.

- Data Grouping: Groups the data in a DataFrame `df` based on the `Group` column. Selects the `Value` column from each group, creating a series of grouped values.
- Box Plot Creation: Creates a boxplot on the `ax` axes. Uses data from the grouped values to form the boxes. Labels each box with the unique group names from the `Group` column.

```python
# Group our dataset with our 'Group' variable
grouped = df.groupby('Group', observed=False)['Value']

# Init a figure and axes
fig, ax = plt.subplots(figsize=(8,6))

# Create the plot
ax.boxplot(x=[group.values for name, group in grouped],
           labels=grouped.groups.keys())

# Display it
plt.show()
```

# 3. Jitter plot and beeswarm plot
As for jitter plot and beeswarm plot, we will use only Seaborn package. As already explained in the previous sections, Seaborn provides higher level functions (more convenient, but might not be able to edit details) compared with Matplotlib, and thus, Seaborn may be more suitable to get an overview of various plots, especially for beginners.

### Jitter plot
- A jitter plot is used to visualize the distribution of many individual values.
- 
The values are plotted as points along one axis, and the dots are then shifted randomly along the other axis, allowing the dots not to overlap


```python
# Use titanic data we loaded previously
titanic

# set groupby(observed = False) to avoid unnecessary warnings
#titanic_sub = titanic[["age", "class"]]
#titanic_sub = titanic_sub.groupby(["age", "class"], observed = False)

# Create a simple jitter plot
sns.stripplot(data = titanic,
              x = "class",
              y = "age",
              jitter = True)
```

If the dots in a jitter plot are not shifted, we may call the plot "stripplot."
Thus, in Seaborn, `stripplot()` is used instead of `jitterplot()`.

We can overlay jitter plots on boxplots using the following code:

```python
# Create boxplots
sns.boxplot(data = titanic,
            x = "class",
            y = "age",
            color = "lightgray")

# Overlay jitter plots
sns.stripplot(data = titanic,
              x = "class",
              y = "age",
              jitter = True)
```

## beeswarm plot
