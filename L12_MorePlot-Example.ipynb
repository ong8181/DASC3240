# 1. Sunspot time series: Load the data and visualize the time series of the monthly sunspot patterns
- By default, the data type of `date` in `sunspot_month_L12.csv` is `pandas.core.series.Series`. You should let your Python know that `date` is a date variable. The `pandas` package also has a convenient function to do this.

````python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import data
d1 = pd.read_csv("data/sunspot_month_L12.csv")
#type(ｄ１["date"])
d1["date"] = pd.to_datetime(d１["date"])

# In this case, "time" indicates dicimal year (= yyyy.xx) (Numeric is OK)
# so we do not need to convert it to datetime object at least for visualization purpose.
plt.figure(figsize=(6,3))
sns.lineplot(data = d1, x = "date", y = "value", linewidth = 0.5)
plt.xlabel("Date")
plt.ylabel("The number of monthly sunspots");
````

## 2. Sunspot time series:  It is known that the number of sunspots fluctuates in an 11-year cycle. Visualize this by drawing vertical lines to the graph every 11 years.
- You can use `pandas.date_range` function to create a sequence of date variables (see [official help document](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.date_range.html)). The option `freq` and [Offset aliases](https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#timeseries-offset-aliases) can be used to specify the time interval.

````python
# Generate time series plot
plt.figure(figsize=(6,3))
sns.lineplot(data = d1, x = "date", y = "value", linewidth = 0.5)
plt.xlabel("Date")
plt.ylabel("The number of monthly sunspots")

# Add vertical lines
date_interval = pd.date_range(start='1749-01-01', end='2014-09-01', freq='11Y')
for vline in date_interval: plt.axvline(vline, color='red', linestyle='--', linewidth = 0.5)
plt.show()
````
