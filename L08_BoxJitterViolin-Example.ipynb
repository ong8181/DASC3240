# `ChickWeight`

## 1. Extract data with `Time == 10` and visualize the effect of `Diet` on `weight` using boxplots overlaid with jitter plots.  
Please be careful of outlier points.

````python
# Import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import data
## ChickWeight data
d = pd.read_csv("data/ChickWeight.csv")

# Compile data
d_sub1 = d[d['Time'] == 10]
#d1_sub1

# Create figure objects
fig1, ax1 = plt.subplots()

# Create boxplots
sns.boxplot(data = d_sub1, x = "Diet", y = "weight", color = "lightgray", showfliers=False)
# Overlay jitter plots
sns.stripplot(data = d_sub1, x = "Diet", y = "weight", jitter = True)
ax1.set_xlabel('Diet')
ax1.set_ylabel('Weight (g)')
ax1.set_title('Chick Weight at Time = 10');
````

## 2. Choose data with `Time == 20` and visualize the effect of `Diet` on `weight` using boxplots overlaid with jitter plots.
Please be careful of outlier points.

````python
# Compile data
d_sub2 = d[d['Time'] == 20]

# Create figure objects
fig2, ax2 = plt.subplots()

# Create boxplots
sns.boxplot(data = d_sub2, x = "Diet", y = "weight", color = "lightgray", showfliers=False)
# Overlay jitter plots
sns.stripplot(data = d_sub2, x = "Diet", y = "weight", jitter = True)
ax2.set_xlabel('Diet')
ax2.set_ylabel('Weight (g)')
ax2.set_title('Chick Weight at Time = 20');
````

## 3. Combine the two figures generated in Task 1 and Task 2 so that readers can easily understand the effects of `Diet` and `Time` on `weight`.  

````python
# Compile data
d_sub3 = d[(d['Time'] == 10) | (d['Time'] == 20)]

# Create figure objects
fig3, ax3 = plt.subplots()

# Set colors
box_colors = ["lightgray", "gray"]
point_colors = ["white", "gray"]

# Create boxplots
sns.boxplot(data = d_sub3, x = "Diet", y = "weight", hue = "Time",
            palette = box_colors, color = "lightgray", showfliers = False, width = 0.7)
# Overlay jitter plots
sns.stripplot(data = d_sub3, x = "Diet", y = "weight", hue = "Time",
              palette = point_colors, edgecolor = "black", linewidth = 1, jitter = True, dodge = True)
ax3.set_xlabel('Diet')
ax3.set_ylabel('Weight (g)')
ax3.set_title('Chick Weight at Time = 10 and 20')
ax3.legend(loc = 'upper left', bbox_to_anchor=(1, 1));
````

# Custom datasets

## 1. Combine the custom datasets (`d1`, `d2`, `d3`, `d4`, and `d5`) into one `pandas.DataFrame`. The five datasets can be downloaded from [here](https://github.com/ong8181/DASC3240/tree/main/data). Visualize them using boxplots overlaied with beeswarm plots (an example is provided below).
When conbiming different datasets, you should remember that you will use the dataframe to create boxplots and beeswarm plots later. In other words, what type of dataframe structure is convenient to create the plots?
- HINT1: It is better to assign one data in one row (so-called "[tidy data](https://cran.r-project.org/web/packages/tidyr/vignettes/tidy-data.html#:~:text=Tidy%20data%20is%20a%20standard,Every%20row%20is%20an%20observation.)").
- HINT2: You should add a new column that indicates the name of the dataset.

````python
# ------------------------------- #
# Combine datasets
# ------------------------------- #
## Load custom datasets
cd1 = pd.read_csv("data/d1.csv")
cd2 = pd.read_csv("data/d2.csv")
cd3 = pd.read_csv("data/d3.csv")
cd4 = pd.read_csv("data/d4.csv")
cd5 = pd.read_csv("data/d5.csv")

## Change column names
cd1 = cd1.rename(columns={'d1': 'value'})
cd2 = cd2.rename(columns={'d2': 'value'})
cd3 = cd3.rename(columns={'d3': 'value'})
cd4 = cd4.rename(columns={'d4': 'value'})
cd5 = cd5.rename(columns={'d5': 'value'})

## Add a new column so that we can create plots easily
cd1.insert(2, "dataset", "d1", True)
cd2.insert(2, "dataset", "d2", True)
cd3.insert(2, "dataset", "d3", True)
cd4.insert(2, "dataset", "d4", True)
cd5.insert(2, "dataset", "d5", True)

## Combine all data
cd = pd.concat([cd1, cd2, cd3, cd4, cd5], axis = 0) 


# ------------------------------- #
# Generate plots
# ------------------------------- #
# Create figure objects
fig4, ax4 = plt.subplots(figsize = [6, 5])

# Boxplot overlaied with beeswarm plot
sns.boxplot(x = "dataset", y = "value", data = cd, color = "white")
sns.swarmplot(x = "dataset", y = "value", data = cd, color = "gray")
ax4.set_xlabel('Dataset name')
ax4.set_ylabel('Value')
ax4.set_title('Boxplot and beaswamplot: How different data look similar if you use boxplots only');
````
