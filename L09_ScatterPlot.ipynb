# 1. Scatter plot with Matplotlib
- **Scatter plot** displays the relationship between two numeric variables, one being displayed on the X-axis (horizontal) and the other on the Y-axis (vertical).


### A basic scatter plot
This example shows how to draw a scatter plot using Matplotlib. This is a basic scatterplot example made with the `plot()` function of Matplotlib. These arguments are passed to the function:

- `x` : The x coordinates of the bars. (`x_pos` in the example)
- `height` : The height(s) of the bars.
- `data` : An object with labelled data
- `linestyle` : Style of the lines between each point
- `marker` : Marker style of the points


(Source: https://python-graph-gallery.com/130-basic-matplotlib-scatterplot/)

```python
# libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 
# Create a dataset:
df=pd.DataFrame({'x_values': range(1,101), 'y_values': np.random.randn(100)*15+range(1,101) })
 
# plot
plt.plot( 'x_values', 'y_values', data=df, linestyle='none', marker='o')
plt.show()
```

### Adding regression line
This guide shows how to plot a scatterplot with an overlaied regression line in Matplotlib. The linear regression fit is obtained with `numpy.polyfit(x, y)` where `x` and `y` are two one dimensional numpy arrays that contain the data shown in the scatterplot. The slope and intercept returned by this function are used to plot the regression line.

```python
# Import libraries (if you have not done yet) and generate random values
#import matplotlib.pyplot as plt
#import numpy as np
rng = np.random.default_rng(1234)
```

Then, let's plot the random values and add a liner regression line to the plot

```python
# Generate data
x = rng.uniform(0, 10, size=100)
y = x + rng.normal(size=100)

# Initialize layout
fig, ax = plt.subplots(figsize = (9, 9))

# Add scatterplot
ax.scatter(x, y, s=60, alpha=0.7, edgecolors="k")

# Fit linear regression via least squares with numpy.polyfit
# It returns an slope (b) and intercept (a)
# deg=1 means linear fit (i.e. polynomial of degree 1)
b, a = np.polyfit(x, y, deg=1)

# Create sequence of 100 numbers from 0 to 100 
xseq = np.linspace(0, 10, num=100)

# Plot regression line
ax.plot(xseq, a + b * xseq, color="k", lw=2.5);
```

## Challanges
1. You can check the contents of `a` and `b` (i.e., slope and intercept). Also, you can change the equation to generate `y` (`y = x + rng.normal(size=100)`), repeat the analysis, and check how `a` and `b` change.
2. Change the `degree` to `2` in `np.polyfit()` and check the new regression line (or curve). You can check the help page (https://numpy.org/doc/stable/reference/generated/numpy.polyfit.html).

# 2. Scatter plot with seaborn
### A basic scatter plot
A scatterplot can be made using `regplot()` function of seaborn library. An example dataset from seaborn repository, `iris` dataset, is used in the example. The plot shows the relationship between sepal length and width of plants. In order to show the most basic utilization of this function, the following parameters should be provided:
- `x` : positions of points on the X axis
- `y` : positions of points on the Y axis


By default, a linear regression fit is plotted. It can be removed from graph by setting `fit_reg = False`.

```python
# Load library and dataset
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('iris')
 
# Use the function regplot to make a scatterplot
sns.regplot(x = df["sepal_length"], y = df["sepal_width"], fit_reg = False)
```

### Adding a linear regression line
By default, a linear regression fit is plotted, so you remove `fit_reg = False`, or set `fit_reg = True` to add a linear regression line.

```python
# Use the function regplot to make a scatterplot with a linear regression line
sns.regplot(x = df["sepal_length"], y = df["sepal_width"], fit_reg = True)
```

Types of regression line/curve and whether confidence intervals should be drawn can be modified using the following paramters.

-  `fit_regbool`: If `True`, estimate and plot a regression model relating the `x` and `y` variables.
-  `ci`: Size of the confidence interval for the regression estimate (e.g., `0.95`).
-  `lowess`: If `True`, estimate a nonparametric lowess model (locally weighted linear regression). Note that confidence intervals cannot currently be drawn for this kind of model.


```python
## Regression line without CIs
sns.regplot(x = df["sepal_length"], y = df["sepal_width"], fit_reg = True, ci = None)
```

```python
## Regression curve using LOWESS
sns.regplot(x = df["sepal_length"], y = df["sepal_width"], lowess = True)
```

According to the officieal website, we cannot draw CI when we use `lowess`. We may use a different library to draw confidence intervals for LOWESS fitting. For the other information about `seaborn.regplot()`, you can check https://seaborn.pydata.org/generated/seaborn.regplot.html.

# 3. Combined with other plots
A scatter plot can be used with other plots. These plots are often plotted at the margin, and thus, they can be called as **margin plot(s)**. A margin plot visualize additional information, which is sometimes useful. As a marginal plot, histogram or denstity plot is often used, but you may use other types of plots such as a boxplot.

The seaborn library provides a joint plot function that is really handy to make this type of graphics. The top graph shows its default behaviour, and here are few possible customizations.

Seaborn library have `jointplot()` function to draw a marginal plot. The kind of the central plot can be given as a parameter in `jointplot()` function:

- `kind` : the possible options are `“scatter”`, `“kde”`, `“hist”`, `“hex”`, `“reg”`, `“resid”`

```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
 
# Custom the inside plot: options are: “scatter” | “reg” | “resid” | “kde” | “hex”
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='scatter')
```

Here are some methods to customize the plot.

```python
# Then you can pass arguments to each type:
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='scatter', s=200, color='m', edgecolor="skyblue", linewidth=2)
```

## 4. Other tips 
### Tips to avoid overplotting so many data points
**Overplotting** is one of the most common problems in data visualization. When your dataset is big, points of your scatterplot tend to overlap, and your graphic becomes unreadable.

To avoid the problem of overplotting, you can use some tips:
- Dot size
- Transparency
- 2D Density plot

```python
# Generate artificial data
# libraries
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

# Dataset:
tmp1 = pd.DataFrame({'x': np.random.normal(10, 1.2, 20000), 'y': np.random.normal(10, 1.2, 20000), 'group': np.repeat('A',20000) })
tmp2 = pd.DataFrame({'x': np.random.normal(14.5, 1.2, 20000), 'y': np.random.normal(14.5, 1.2, 20000), 'group': np.repeat('B',20000) })
tmp3 = pd.DataFrame({'x': np.random.normal(9.5, 1.5, 20000), 'y': np.random.normal(15.5, 1.5, 20000), 'group': np.repeat('C',20000) })
df2 = pd.concat([tmp1, tmp2, tmp3], axis=0)
```

First, we use a default scatter plot using `seaborn.scatterplot()`.

```python
# Overplotting figure
# Plot with small marker size
sns.scatterplot(x = 'x', y = 'y', data = df2)
```

Now, let's try to change the dot size.

```python
# Dot size
# Plot with small marker size
sns.scatterplot(x = 'x', y = 'y', data = df2, s = 0.5)
```

Changing the transparency of the ponts is also an option.

```python
# Transparency
# Plot with smaller alpha (higher transparency)
sns.scatterplot(x = 'x', y = 'y', data = df2, alpha = 0.01)
```

2D density plot is also good (This may take some time).

```python
sns.kdeplot(x = 'x', y = 'y', data = df2, fill = True, cmap = 'Reds')
```
