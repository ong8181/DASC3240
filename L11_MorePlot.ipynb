## Warm-up: `datetime` object
You can create a `datetime` object using the `datetime` class. Here's an example:

```python
# libraries
import datetime
 
# create a datetime object for the current date and time
now = datetime.datetime.now()
print(now)
```

```python
# Check the type of now
type(now)
```

### Changing the format
You can change the format of a datetime object using the `strftime()` method.

```python
# format the datetime object as a string
formatted_date = now.strftime("%Y-%m-%d %H:%M:%S")
print(formatted_date)
```

### Creating a datetime object from a string
You can also create a datetime object from a string using the `strptime()` method.

```python
# create a datetime object from a string
date_string1 = "2021-09-08 16:32:45"
date_object1 = datetime.datetime.strptime(date_string1, "%Y-%m-%d %H:%M:%S")

print(date_object1)
```

```python
# create a datetime object from a string
date_string2 = "2021-09-08"
date_object2 = datetime.datetime.strptime(date_string2, "%Y-%m-%d")

print(date_object2)
```

```python
# Calculate the difference between two dates
print(date_object1 - now)
```

## Plotting time series
You can use `seaborn.lineplot()` to generate a time series plot. You need to make sure that the x-axis is "date", NOT "character". (If you want to use `matplotlib`, you can simply use `plt.plot()`).

```python
# Load libraries
import pandas as pd
import seaborn as sns

# Load data from web
data = pd.read_csv(
    "https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/3_TwoNumOrdered.csv", 
    delim_whitespace=True
)

# Let's make sure 'date' is actually a date in pandas
date = pd.to_datetime(data["date"]) # This method is provided by the 'pandas' package
#date = data["date"] # What happens if you do this?
value = data["value"]

# Plot a time series line plot with Seaborn
plt.figure(figsize=(5, 3))
sns.lineplot(x = date, y = value);
```

### Customize a time series plot
To add markders (symbols) to a time series plot, you can set the `markers` True. (Note that the data is too dense and the plot is "overplotting")

```python
# plot a time series line plot with markers
fig, ax = plt.subplots(2, 2, figsize=(10, 8))
ax = ax.ravel()

# plot a time series line plot with markers
sns.lineplot(x = date, y = value, markers = True, marker = "D", ax = ax[0])
ax[0].set_title("With markers")
# Reduce the marker edge widths
sns.lineplot(x = date, y = value, markers = True, marker = "D", markeredgewidth = 0.1, ax = ax[1])
ax[1].set_title("Reduce marker edge widths")
# Change the marker style
sns.lineplot(x = date, y = value, markers = True, marker = "o", markeredgewidth = 0.1, ax = ax[2])
ax[2].set_title("Change marker style")
# Change the marker size and line width
sns.lineplot(x = date, y = value, markers = True, marker = "o", markeredgewidth = 0.1, markersize = 5, linewidth = 1, ax = ax[3])
ax[3].set_title("Change marker size and line width")

# Adjust the intervals between figures
plt.tight_layout()
```

# 2. Dendrogram
A **dendrogram** is a network structure. It is constituted of a root node that gives birth to several nodes connected by edges or branches.


To draw a dendrogram, you first need to have a **numeric matrix** (or if we can calculate a **distance matrix** from the data, e.g., sequence data, it is fine). Each line represents an entity (here a car). Each column is a variable that describes the cars. The objective is to cluster the entities to show who shares similarities with whom. The dendrogram will draw the similar entities closer to each other in the tree.


Let’s start by loading a dataset and the requested libraries:


```python
# Libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
 
# Import the mtcars dataset from the web + keep only numeric variables
url = "https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/mtcars.csv"
#url = "https://vincentarelbundock.github.io/Rdatasets/csv/datasets/mtcars.csv"
df = pd.read_csv(url)
df = df.set_index("model")
#df = df.reset_index(drop=True)
df.head()
```

All right, now that we have our numeric matrix, we can calculate the distance between each car, and draw the hierarchical clustering. Distance calculation can be done by the `linkage()` function.

```python
# Calculate the distance between each sample
# You have to think about which metric you use (how to measure distance/similarity) + about which clustering method you use (How to group cars)
Z = linkage(df, "ward")
```

Let's briefly check what is inside Z.

```python
Z
```

Z is a numerix matrix. Then, you can easily plot this object as a dendrogram using the `dendrogram()` function of `scipy` library. These parameters are passed to the function:

- `Z` : The linkage matrix
- `labels` : Labels to put under the leaf node
- `leaf_rotation` : Specifies the angle (in degrees) to rotate the leaf labels


```python
# Plot title
plt.title("Hierarchical Clustering Dendrogram")

# Plot axis labels
plt.xlabel("sample index")
plt.ylabel("distance (Ward)")

# Make the dendrogram
dendrogram(Z, labels=df.index, leaf_rotation=90)

# Show the graph
plt.show()
```

# 3. Network
A **network** shows interconnections between a set of entities. Each entity is represented by a node. Connection between nodes are represented through links (or edges).

A network chart is constituted by nodes. These nodes are interconnected by edges. So a basic format is a **data frame** where each line describes a connection. *Note that the connections are already given in this example, so it is different from a correlation-based network reconstruction I explained in the lecture.*

Here we construct a data frame with four lines, describing the four connections of this plot! So if you have a csv file with your connections, load it and you are ready to visualise it!

Next step: customise the chart parameters!  
(Need to install `networkx` if you have not; e.g. use `conda install anaconda::networkx` in Terminal)


```python
# libraries
#import pandas as pd
#import numpy as np
import networkx as nx  # Need to install this package if you have not
import matplotlib.pyplot as plt
 
# Build a dataframe with 4 connections
df_net = pd.DataFrame({ "from":["A","B","C","A"], "to":["D","A","E","C"]})

# Check the structure of the dataframe
df_net
```

```python
# Build your graph
G = nx.from_pandas_edgelist(df_net, "from", "to")
 
# Plot it
nx.draw(G, with_labels=True)
plt.show()
```

Then, let's customize the network.


https://python-graph-gallery.com/321-custom-networkx-graph-appearance/

### Customize network

```python
# plot a time series line plot with markers
fig2, ax2 = plt.subplots(2, 2, figsize=(10, 8))
ax2 = ax2.ravel()

# Graph with Custom nodes
nx.draw(G, with_labels=True, node_size=150, node_color="skyblue", node_shape="s", alpha=0.5, linewidths=10, ax=ax2[0])
ax2[0].set_title("Custom nodes")
# Custom the labels
nx.draw(G, with_labels=True, node_size=1500, font_size=25, font_color="yellow", font_weight="bold", ax=ax2[1])
ax2[1].set_title("Custom labels")
# Chart with Custom edges
nx.draw(G, with_labels=True, width=5, edge_color="skyblue", style="solid", ax=ax2[2])
ax2[2].set_title("Custom edges")
# All together we can do something fancy
nx.draw(G, with_labels=True, node_size=1500, node_color="skyblue", node_shape="o", alpha=0.5, linewidths=4, font_size=25, 
        font_color="grey", font_weight="bold", width=2, edge_color="grey", ax=ax2[3])
ax2[3].set_title("All together")

# Show figures
plt.tight_layout()
```

# 4. Map

A map is basically a set of polygons displayed on a two dimensional canvas. There are two problems you need to solve when building one:

- Where to find polygon coordinates. It can come from:
    - a **shape file** or a **geoJson** file. That's two different ways to store the information. The `geopandas` package is the best way to read those file formats with Python.
    - a **python library** like `geopandas` or `basemap` providing the information for common location like the US.
    - **open street map** (ODbL 1.0) and **google map** (need to care about license issues...?).  
- How to plot those polygons?
    - Once you get a `geodataframe` thanks to the `geopandas` package, `geoplot` is your best choice to build a static map
    - If you need an interactive map from a `geodataframe`, `plotly` is a good option.
    - If you want to use google map style maps, `folium` is the way to go.

(Source: https://python-graph-gallery.com/map/)

### Random walk v.s. random values

```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Generate random walk
n_steps = 1000
walk = np.zeros(n_steps)
for i in range(1, n_steps):
    walk[i] = walk[i-1] + np.random.randn()

## Random walk (one-liner)
# walk = np.cumsum(np.random.randn(n_steps))

# Generate random values
n_values = 1000
values = np.random.randn(n_values)

# Create figure and axes objects
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8))

# Plot random walk
ax1.plot(walk)
ax1.set_title("Random Walk")
ax1.set_xlabel("Time"); ax1.set_ylabel("Value")

# Plot random values
ax2.plot(values)
ax2.set_title("Random Values")
ax2.set_xlabel("Time"); ax2.set_ylabel("Value")

# Adjust layout and display plot
fig.tight_layout()
plt.show()
```
