### Preparation

```python
# Import pandas
import pandas as pd
# Import matplotlib and seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Import data
d1 = pd.read_csv("data/ChickWeight.csv")
d2 = pd.read_csv("data/chickwts.csv")
d3 = pd.read_csv("data/InsectSprays.csv")
```

# `ChickWeight`

### 1. Create a simple histogram of `weight`. Please use all data. 

```python
# Draw histogram
fig1, ax1 = plt.subplots()
sns.histplot(data=d1, x='weight', bins=20, kde=True)
ax1.set_xlabel('Weight (g)')
ax1.set_ylabel('Frequency')
ax1.set_title('Histogram of Chick Weight');
```

### 2. Exclude data with `Time < 10` and create histogram of `weight` for each `Diet`. There are four `Diet`, so you will have four histograms, but please overlay all the histograms in one figure.

```python
# Exclude data with Time < 10 and create histograms of weight for each Diet
d1_sub = d1[d1['Time'] >= 10]

# Draw histogram
fig2, ax2 = plt.subplots()
sns.histplot(data=d1_sub, x='weight', hue='Diet', bins=20, kde=True)
ax2.set_xlabel('Weight (g)')
ax2.set_ylabel('Frequency')
ax2.set_title('Effects of Diet on Chick Weight');
```

### 3. Draw each histogram in each separate penel (facet) and combine them in a 2 x 2 format. You will have four facets in one large figure.

```python
# Create four data subsets
d1_sub1 = d1_sub[d1_sub["Diet"] == 1]
d1_sub2 = d1_sub[d1_sub["Diet"] == 2]
d1_sub3 = d1_sub[d1_sub["Diet"] == 3]
d1_sub4 = d1_sub[d1_sub["Diet"] == 4]

# Set bin width and x-value range
bin_width = 10

# Adjust appearance
fig3, ax3 = plt.subplots(2, 2, figsize=(8, 6))
fig3.suptitle('Effects of Diet on Chick Weight', fontsize=16) # Set the title for all four histograms

# Adjust appearance
## Panel 1
ax3[0,0].hist(data=d1_sub1, x='weight', bins=range(40, 210, bin_width), edgecolor='black')
ax3[0,0].set_ylim(0, 14)  # Set the y-axis range
ax3[0,0].set_xlabel('Weight (g)')
ax3[0,0].set_ylabel('Frequency')

## Panel 2
ax3[0,1].hist(data=d1_sub2, x='weight', bins=range(40, 210, bin_width), edgecolor='black')
ax3[0,1].set_ylim(0, 14)  # Set the y-axis range
ax3[0,1].set_xlabel('Weight (g)')
ax3[0,1].set_ylabel('Frequency')

## Panel 3
ax3[1,0].hist(data=d1_sub3, x='weight', bins=range(40, 210, bin_width), edgecolor='black')
ax3[1,0].set_ylim(0, 14)  # Set the y-axis range
ax3[1,0].set_xlabel('Weight (g)')
ax3[1,0].set_ylabel('Frequency')

## Panel 4
ax3[1,1].hist(data=d1_sub4, x='weight', bins=range(40, 210, bin_width), edgecolor='black')
ax3[1,1].set_ylim(0, 14)  # Set the y-axis range
ax3[1,1].set_xlabel('Weight (g)')
ax3[1,1].set_ylabel('Frequency')

plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.show()
```
